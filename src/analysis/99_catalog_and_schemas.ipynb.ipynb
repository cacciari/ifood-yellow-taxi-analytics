# 📂 Data Catalog & Schema Setup
---
This notebook initializes the **Medallion Architecture** environment for the iFood Data Quality Challenge. It establishes the catalog, logical schemas, and core table structures.

### Architecture Blueprint
* **Lake**: Lake schema where everything goes.
    * **Landing Zone**: Landing Zone Volume specifically for NYX TLC Data
* **Silver**: Conformed layer containing validated, typed, and deduplicated data.
* **Gold**: Aggregated business metrics optimized for downstream analytics.


### Catalog and Schemas SQL Statements
 

```sql
-- General Catalog for this presentation
CREATE CATALOG IF NOT EXISTS IFOOD
COMMENT "Catalog for iFood Data Quality Engineer challenge -- MEDALLION";
USE CATALOG IFOOD;

-- Dedicated schema do govern our "lake"
CREATE SCHEMA IF NOT EXISTS LAKE
COMMENT "Our 'Lake' schema, where we ingest our raw data";

-- Landing Zone
CREATE VOLUME IF NOT EXISTS LAKE.LANDING_ZONE
COMMENT "Landing Zone for Parquet files ingested from NYC TLC repos";

-- BRONZE LAYER FOR RAW DATA
CREATE SCHEMA IF NOT EXISTS BRONZE
COMMENT "Data aquisition from LAKE"; 

-- SILVER LAYER FOR QUALITY CONTROL
CREATE SCHEMA IF NOT EXISTS SILVER
COMMENT "Cleaned and standardized data";

-- GOLD LAYER FOR ANALYTICS
CREATE SCHEMA IF NOT EXISTS GOLD
COMMENT "Data ready for analytics";
```
